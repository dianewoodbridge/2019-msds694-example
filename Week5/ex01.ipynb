```python
import pyspark
```

```python
sc = pyspark.SparkContext.getOrCreate()
```

```python
numbers = sc.textFile("../Data/numbers.txt", 5).map(lambda x : int(x))
```

## Sort numbers in each partition

```python
numbers.glom().take(1)
```

```python
numbers.mapPartitions(lambda x : sorted(x)).glom().take(1)
```

```python
numbers.map(lambda x : sorted(x)).glom().take(1)
```

```python
numbers.mapPartitions(lambda x : sorted(x)).glom().take(1)
```

```python
def sum_map_map_partition(iterator): 
    yield sum(iterator)
```

```python
numbers.mapPartitions(sum_map_map_partition).glom().collect()
```

## What is the difference between .map() and .mapPartitions()

```python
import time
class Power:
    def __init__(self, p):
        self.p = p
        time.sleep(2)
        
    def applyPower(self, x):
        return x**self.p
        
# mapPartitions        
def power_map_partitions(nums):
    c = Power(5)
    
    for x in nums:
        yield c.applyPower(x)
        
# map
def power_map(num):
    c = Power(5)
    return c.applyPower(num)
```

```python
t = time.time()
numbers.map(power_map).collect()
time.time() - t
```

```python
t = time.time()
numbers.mapPartitions(power_map_partitions).collect()
time.time() - t
```
